# Cumulative Distribution Functions

The CDF is a function of x, where x is any value that might appear in the distribution. To evaluate  for a particular value of x, we compute the fraction of values in the distribution less than or equal to x.

Here’s what that looks like as a function that takes a sequence, t, and a value, x:

```python
def EvalCdf(t, x):
    count = 0.0
    for value in t:
        if value <= x:
            count += 1

    prob = count / len(t)
    return prob

list_t = [1, 2, 2, 3, 5]
EvalCdf(list_t, 5)
```

# Hypothetical Situation: Online Video Streaming Service

Imagine you are working for an online video streaming service, and your team is responsible for optimizing the video buffering time for users. You want to model the time intervals between consecutive video requests from a user to estimate how frequently users request new videos.

# Implementation:

Here's how you can use the numpy.random.exponential function to simulate this situation:
# New Section
Import Necessary Libraries:
You'll need to import the numpy library to generate random numbers from the exponential distribution.

```python
import numpy as np
import matplotlib.pyplot as plt
```

# Set Parameters:
Define the scale parameter (β), which represents the average time between video requests. Let's say the average time between requests is 30 seconds, so β=30

```python
beta = 30  # Average time between video requests in seconds
```

# Generate Random Time Intervals:
Use the numpy.random.exponential function to generate random time intervals between video requests for a user. You can specify the number of intervals you want to generate using the size parameter.

```python
num_intervals = 100  # Number of time intervals to generate
time_intervals = np.random.exponential(scale=beta, size=num_intervals)
```

 time_intervals will contain an array of 10 random time intervals between video requests, following an exponential distribution with the specified scale parameter (β).

# Simulate User Behavior:
You can use these generated time intervals to simulate a user's behavior in requesting videos. For example, you could create a loop that simulates video requests over time based on the generated time intervals.

```python
video_start_time = 0  # Initial time
for interval in time_intervals:
    # Simulate video request at the next time interval
    video_start_time += interval
    # print(f"User requested a video at time {video_start_time} seconds.")
```

This loop simulates when a user requests videos based on the generated time intervals. You can use this information to analyze user behavior and optimize your video streaming service accordingly.

# Plot a histogram of the time intervals

```python
plt.hist(time_intervals, bins=10, density=True, alpha=0.6, color='g', label='Histogram')

# Calculate the PDF using the formula
x = np.linspace(0, max(time_intervals), 100)
pdf = (1 / beta) * np.exp(-x / beta)

# Plot the PDF
plt.plot(x, pdf, 'r-', lw=2, label='PDF')

# Add labels and legend
plt.xlabel('Time between Video Requests (seconds)')
plt.ylabel('Probability Density')
plt.legend()


# Show the plot
plt.title('Exponential Distribution PDF for Video Request Intervals')
plt.grid()
plt.show()
```
